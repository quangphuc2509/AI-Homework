```python
import pandas as pd 
import numpy  as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
url = 'lung_cancer_examples.csv'
data = pd.read_csv(url)
data.hist()
plt.show

correlation = data.corr()
print(correlation)
```

```python
sns.heatmap(correlation, annot = True)
sns.pairplot(data)
```

```python
data.plot(kind = 'scatter', x = 'Alkhol', y = 'Result', figsize=(9,9), color = 'black')
```

```python
#M hnh bc 1
modeldata1 = linear_model.LinearRegression()
X = pd.DataFrame(data['Alkhol'])
modeldata1.fit(X, y = data['Result'])
print('Intercept ca m hnh: ', modeldata1.intercept_)
print('Coeffient ca m hnh: ', modeldata1.coef_)
print(' chnh xc ca m hnh: ', modeldata1.score(X,y = data['Result']))
data.plot(kind = 'scatter', x = 'Alkhol', y = 'Result', figsize=(9,9), color = 'black')
y_pred = modeldata1.predict(X)
plt.plot(X,y_pred, color = 'blue')
```

```python
#M hnh bc 5
modeldata5 = linear_model.LinearRegression()
X = pd.DataFrame([data['Alkhol'], data['Alkhol']**2, data['Alkhol']**3, data['Alkhol']**4,data['Alkhol']**5]).T
modeldata5.fit(X, y = data['Result'])
print('Intercept ca m hnh: ', modeldata5.intercept_)
print('Coeffient ca m hnh: ', modeldata5.coef_)
print(' chnh xc ca m hnh: ', modeldata5.score(X,y = data['Result']))

import numpy as np
range = np.arange(data['Alkhol'].min(),data['Alkhol'].max(),0.1)
data.plot(kind = 'scatter', x = 'Alkhol', y = 'Result', figsize=(9,9),color = 'black')
predictor = pd.DataFrame([range, range**2, range**3, range**4, range**5]).T
y_pred = modeldata5.predict(X = predictor)
plt.plot(range,y_pred, color = 'blue')
```

```python
#M hnh bc 12
modeldata12 = linear_model.LinearRegression()
X = pd.DataFrame([data['Alkhol'], data['Alkhol']**2, data['Alkhol']**3, data['Alkhol']**4,data['Alkhol']**5, data['Alkhol']**6, 
                  data['Alkhol']**7, data['Alkhol']**8, data['Alkhol']**9, data['Alkhol']**10, data['Alkhol']**11, data['Alkhol']**12]).T
modeldata12.fit(X, y = data['Result'])
print('Intercept ca m hnh: ', modeldata12.intercept_)
print('Coeffient ca m hnh: ', modeldata12.coef_)
print(' chnh xc ca m hnh: ', modeldata12.score(X,y = data['Result']))

import numpy as np
range = np.arange(data['Alkhol'].min(),data['Alkhol'].max(),0.1)
data.plot(kind = 'scatter', x = 'Alkhol', y = 'Result', figsize=(9,9),color = 'black')
predictor = pd.DataFrame([range, range**2, range**3, range**4, range**5, range**6, range**7, range**8, range**9, range**10, range**11, range**12]).T
y_pred = modeldata12.predict(X = predictor)
plt.plot(range,y_pred, color = 'blue')
```
