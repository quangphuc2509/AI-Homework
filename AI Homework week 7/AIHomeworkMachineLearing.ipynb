```python
import pandas as pd 
import numpy  as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
url = 'Housing.csv'
data = pd.read_csv(url)
data.hist()
plt.show

correlation = data.corr()
print(correlation)
```

```python
sns.heatmap(correlation, annot = True)
sns.pairplot(data)
```

```python
data.plot(kind = 'scatter', x = 'area', y = 'price', figsize = (9,9), color = 'black')
```

```python
# Mô hình bậc 1
model = linear_model.LinearRegression()
X = pd.DataFrame([data['area']]).T
y = data['price']
model.fit(X, y)
print('Intercept của mô hình: ',model.intercept_)
print('Coeffient của mô hình: ',model.coef_)
print('Độ chính xác của mô hình: ', model.score(X, y))
```

```python
data.plot(kind = 'scatter', x = 'area', y = 'price', figsize = (9,9), color = 'black')
y_pred = model.predict(X = pd.DataFrame(data['area']))
plt.plot(data['area'], y_pred, color = 'blue')
```

```python
# Mô hình bậc 3
model3 = linear_model.LinearRegression()
X = pd.DataFrame([data['area'], data['area']**2, data['area']**3]).T
y = data['price']
model3.fit(X, y)
print('Intercept của mô hình: ',model3.intercept_)
print('Coeffient của mô hình: ',model3.coef_)
print('Độ chính xác của mô hình: ', model3.score(X, y))
```

```python
range = np.arange(data['area'].min(),data['area'].max(),1)
predictor = pd.DataFrame([range, range**2, range**3]).T
data.plot(kind = 'scatter', x = 'area', y = 'price', figsize = (9,9), color = 'black')
y_pred = model3.predict(X = predictor)
plt.plot(range, y_pred, color = 'blue')
```

```python
# Mô hình bậc 5
model5 = linear_model.LinearRegression()
X = pd.DataFrame([data['area'], data['area']**2, data['area']**3, data['area']**4, data['area']**5]).T
y = data['price']
model5.fit(X, y)
print('Intercept của mô hình: ',model5.intercept_)
print('Coeffient của mô hình: ',model5.coef_)
print('Độ chính xác của mô hình: ', model5.score(X, y))
```

```python
range = np.arange(data['area'].min(),data['area'].max(),1)
predictor = pd.DataFrame([range, range**2, range**3, range**4, range**5]).T
data.plot(kind = 'scatter', x = 'area', y = 'price', figsize = (9,9), color = 'black')
y_pred = model5.predict(X = predictor)
plt.plot(range, y_pred, color = 'blue')
```
