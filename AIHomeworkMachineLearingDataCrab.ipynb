```python
import pandas as pd 
import numpy  as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
url = 'data_crab.csv'
data = pd.read_csv(url)
data.hist()
plt.show

correlation = data.corr()
print(correlation)
```

```python
sns.heatmap(correlation, annot = True)
sns.pairplot(data)
```

```python
data.plot(kind = 'scatter', x = 'BD', y = 'FL', figsize=(9,9), color = 'black')
```

```python
#Mô hình bậc 1
modeldata1 = linear_model.LinearRegression()
X = pd.DataFrame(data['BD'])
modeldata1.fit(X, y = data['FL'])
print('Intercept của mô hình: ', modeldata1.intercept_)
print('Coeffient của mô hình: ', modeldata1.coef_)
print('Độ chính xác của mô hình: ', modeldata1.score(X,y = data['FL']))
data.plot(kind = 'scatter', x = 'BD', y = 'FL', figsize=(9,9), color = 'black')
y_pred = modeldata1.predict(X)
plt.plot(X,y_pred, color = 'blue')
```

```python
#Mô hình bậc 5
modeldata5 = linear_model.LinearRegression()
X = pd.DataFrame([data['BD'], data['BD']**2, data['BD']**3, data['BD']**4,data['BD']**5]).T
modeldata5.fit(X, y = data['FL'])
print('Intercept của mô hình: ', modeldata5.intercept_)
print('Coeffient của mô hình: ', modeldata5.coef_)
print('Độ chính xác của mô hình: ', modeldata5.score(X,y = data['FL']))

import numpy as np
range = np.arange(data['BD'].min(),data['BD'].max(),0.1)
data.plot(kind = 'scatter', x = 'BD', y = 'FL', figsize=(9,9),
                 color = 'black')
predictor = pd.DataFrame([range, range**2, range**3, range**4, range**5]).T
y_pred = modeldata5.predict(X = predictor)
plt.plot(range,y_pred, color = 'blue')
```
